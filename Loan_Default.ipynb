# Correlation

```python
#Correlation with Target
cr=train.corr()
cr.loss
```

```python
#Low corrleated variables with target

del_cols = [col for col in cr.loss.index if cr.loss[col]<=0.002 and cr.loss[col]>=-0.002 ]
len(del_cols)
```

Remove these columns

```python
data=train.drop(del_cols,axis=1)
data.shape
```

```python
#Data Partition
X=data.drop('loss',axis=1)
y=data['loss']
```

```python
#Multi-Colliniarity
crr=X.corr()
```

```python
#List tightly correlated items

def multi_crr(X):
    crr=X.corr()
    crr_variables=[]
    crr_var=[]
    cols_list=crr.columns.to_list()
    for i in cols_list:
        val=0
        for j in cols_list:
            if i!=j:
                if crr[i][j]>0.8:
                    if ([i,j] not in crr_variables) and ([j,i] not in crr_variables):
                                crr_variables.append([i,j])
                                cols_list.remove(j)
                                
    output = []
    for x in crr_variables:
        if x[1] not in output:
            output.append(x[1])                            
    return output   
```

```python
crr_var = multi_crr(X)
crr_var 
```

Remove above variables

```python
X1=X.drop(crr_var,axis=1)
X1.shape
```

```python
#again test multi colliniarity
mul=multi_crr(X1)
mul
```

Drop these

```python
X2=X1.drop(mul,axis=1)
X2.shape
```

```python
X2.head()
```

```python
#Add the target variable to the X2
X2['Target']=y
X2.head()
```

```python
#Verify columns with single value and drop them 

single_value_cols=[col for col in X2.columns if len(X2[col].unique())==1]
single_value_cols
```

Above variables are not explained target variable .Drop them

```python
X3=X2.drop(single_value_cols,axis=1)
X3.shape
```

```python
#Transform all the variables into numerical
for col in X3.columns:
    X3[col] = pd.to_numeric(X3[col],errors='coerce')
```

```python
from decimal import Decimal

#Scaling the RFM data
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
X_norm = standard_scaler.fit_transform(X3)
X_norm
```

```python
#Data Splitting
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train, y_test = train_test_split(X_norm ,y ,random_state=7)
```

```python
#Random Forest

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,roc_auc_score,roc_curve,accuracy_score



RFC = RandomForestClassifier(n_estimators=100, random_state=1,bootstrap =True,criterion='entropy', max_depth=15 )
RFC.fit(X_train, y_train)
y_pred = RFC.predict(X_test)

print('Accuracy',accuracy_score(y_test,y_pred))
print('Confusion matrix:  \n', confusion_matrix(y_test,y_pred))
```
